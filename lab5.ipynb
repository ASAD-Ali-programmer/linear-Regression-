```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression # Corrected import
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

data1 = {
    'SqFt': [1500, 1800, 2400, 3000, 3500, 4000, 4500, 5000],
    'Price': [300000, 350000, 450000, 500000, 580000, 650000, 720000, 800000]
}
df=pd.DataFrame(data1) # Added pd.DataFrame
X=df[['SqFt']]
y=df['Price']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
model=LinearRegression() # Corrected class name
model.fit(X_train,y_train)

# Corrected the prediction input to be a DataFrame with the feature name 'SqFt'
pred_2800 = model.predict(pd.DataFrame({'SqFt': [2800]}))

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("Predicted price for 2800 SqFt:", pred_2800[0])
print("MAE:", mae)
print("RMSE:", rmse)
print("R2 Score:", r2)

plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Square Footage")
plt.ylabel("Price")
plt.show()
```

```python
# question No 2
data2 = {
    'RD_Spend': [165349, 162597, 153441, 144372, 142107, 131876, 134615],
    'Admin_Spend': [136897, 151377, 101145, 118671, 91391, 99814, 147198],
    'Marketing_Spend': [471784, 443898, 407934, 383199, 366168, 362861, 127716],
    'Profit': [192261, 191792, 191050, 182901, 166187, 156991, 156122]
}

df2 = pd.DataFrame(data2)

X = df2[['RD_Spend','Admin_Spend','Marketing_Spend']]
y = df2['Profit']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

model2 = LinearRegression()
model2.fit(X_train, y_train)

coefficients = model2.coef_
r2 = model2.score(X_test, y_test)

print("Coefficients:", coefficients)
print("R2 Score:", r2)
```
